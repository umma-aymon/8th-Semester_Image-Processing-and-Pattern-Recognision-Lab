```python
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow
from PIL import Image, ImageEnhance 
import matplotlib.pylab as plt
from google.colab import files

uploaded = files.upload()
```

```python
# Read and display the image
image = Image.open('cat.jpg')
#plt.imshow(image)
#Increase the brightness of the image
BrightImage = ImageEnhance.Brightness(image).enhance(2)

#Decrease the brightness of the image
DarkImage = ImageEnhance.Brightness(image).enhance(0.4)

FinalImage = np.concatenate((image, BrightImage, DarkImage), axis=1)
cv2_imshow(cv.cvtColor(FinalImage, cv.COLOR_BGR2RGB))
```
