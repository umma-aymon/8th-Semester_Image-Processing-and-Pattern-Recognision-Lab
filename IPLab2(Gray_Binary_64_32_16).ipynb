# 2. Converting Grayscale to Binary image

```python
img = Image.open('cat.jpg')

gray = img.convert('LA')
binary = gray.convert('1')

f = plt.figure(figsize=(30, 30))

f.add_subplot(1, 3, 1)
plt.title(['Original Image'], fontsize=25)
plt.imshow(img)
plt.axis('off')

f.add_subplot(1, 3, 2)
plt.title(['Gray Image'], fontsize=25)
plt.imshow(gray)
plt.axis('off')

f.add_subplot(1, 3, 3)
plt.title(['Binary Image'], fontsize=25)
plt.imshow(binary)
plt.axis('off')
plt.show(block=True)
```

# **3. Reducing the levels of a grayscale image from 256 to 64, 32 and 16.**

```python
g = np.array(img, dtype=np.uint8)
gray2 = cv.cvtColor(g, cv.COLOR_BGR2GRAY)

f = plt.figure(figsize=(40, 40))

f.add_subplot(3, 1, 1)
ratio = np.amax(gray2) / 64.0
img64 = (gray2 / ratio).astype('uint8')
plt.title(['Level-64'], fontsize=25)
plt.imshow(img64, cmap='gray')
plt.axis('off')

f.add_subplot(3, 1, 2)
ratio = np.amax(gray2) / 32.0 
img32 = (gray2 / ratio).astype('uint8')
plt.title(['Level-32'], fontsize=25)
plt.imshow(img32, cmap='gray')
plt.axis('off')

f.add_subplot(3, 1, 3)
ratio = np.amax(gray2) / 16.0 
img16 = (gray2 / ratio).astype('uint8')
plt.title(['Level-16'], fontsize=25)
plt.imshow(img16, cmap='gray')
plt.axis('off')

plt.show(block=True)
```
