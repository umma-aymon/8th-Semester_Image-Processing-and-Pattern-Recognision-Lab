#Monadic Operations:
1. Adding a constant
2. Subtracting a constant
3. Negative Image
4. Multiplying by a constant
5. Divide by a constant
6. Divide into a constant
7. AND operation
8. OR operation
9. XOR operation



```python
f = plt.figure(figsize=(45, 40))
f.add_subplot(3, 3, 1)
plt.title(['Addition'], fontsize=30)
plt.imshow(g + 50)
plt.axis('off')
f.add_subplot(3, 3, 2)
plt.title(['Subtraction'], fontsize=30)
plt.imshow(g - 50)
plt.axis('off')
f.add_subplot(3, 3, 3)
plt.title(['Negative'], fontsize=30)
plt.imshow(50 - g)
plt.axis('off')
f.add_subplot(3, 3, 4)
plt.title(['Multiplication'], fontsize=30)
plt.imshow(g * 10)
plt.axis('off')
f.add_subplot(3, 3, 5)
plt.title(['Divide By'], fontsize=30)
plt.imshow(g / 10)
plt.axis('off')
f.add_subplot(3, 3, 6)
plt.title(['Divide Into'], fontsize=30)
plt.imshow(10 / g)
plt.axis('off')
f.add_subplot(3, 3, 7)
plt.title(['And'], fontsize=30)
plt.imshow(cv.bitwise_and(g, 1))
plt.axis('off')
f.add_subplot(3, 3, 8)
plt.title(['Or'], fontsize=30)
plt.imshow(cv.bitwise_or(g, 0))
plt.axis('off')
f.add_subplot(3, 3, 9)
plt.title(['Xor'], fontsize=30)
plt.imshow(cv.bitwise_xor(g, 1))
plt.axis('off')
plt.show(block=True)
```
