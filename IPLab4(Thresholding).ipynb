# Single Thresholding (Dark background Ligh object)

```python
g2 = cv.imread('g.jpg', 0)
single = cv.imread('g.jpg', 0)
#print(g2)
[M, N] = single.shape
for i in range(M):
  for j in range(N):
    if single[i,j] < 128:
      single[i,j] = 0         # Dark background Ligh object
    else:
      single[i,j] = 1

f = plt.figure(figsize=(20, 20))
f.add_subplot(1, 2, 1)
plt.title(['Original Grayscale Image'], fontsize=30)
plt.imshow(g2, cmap='gray')
plt.axis('off')
f.add_subplot(1, 2, 2)
plt.title(['Single Thresholding Image'], fontsize=30)
plt.imshow(single, cmap='gray')
plt.axis('off')
plt.show(block=True)
#print(g2)
```

# Dual Thresholding

```python
g2 = cv.imread('g.jpg', 0)
dual = cv.imread('g.jpg', 0)
[M, N] = dual.shape
T1 = 125
T2 = 200
for i in range(M):
  for j in range(N):
    if (not(dual[i,j] >= T1 and dual[i,j] <= T2)):
       dual[i,j] = 0

f = plt.figure(figsize=(20, 20))
f.add_subplot(1, 2, 1)
plt.title(['Original Grayscale Image'], fontsize=30)
plt.imshow(g2, cmap='gray')
plt.axis('off')
f.add_subplot(1, 2, 2)
plt.title(['Dual Thresholding Image'], fontsize=30)
plt.imshow(dual, cmap='gray')
plt.axis('off')
plt.show(block=True)
```

# Grayscale Thresholding

```python
g2 = cv.imread('g.jpg', 0)
grayscale = cv.imread('g.jpg', 0)
[M, N] = grayscale.shape
T1 = 150
T2 = 200
for i in range(M):
  for j in range(N):
    if (grayscale[i,j] >= T1 and grayscale[i,j] <= T2):
       grayscale[i,j] = 255  # white
    else:
       grayscale[i,j] = 0    # black

f = plt.figure(figsize=(20, 20))
f.add_subplot(1, 2, 1)
plt.title(['Original Grayscale Image'], fontsize=30)
plt.imshow(g2, cmap='gray')
plt.axis('off')
f.add_subplot(1, 2, 2)
plt.title(['Grayscale Thresholding Image'], fontsize=30)
plt.imshow(grayscale, cmap='gray')
plt.axis('off')
plt.show(block=True)
```

# Gray Level Slicing: 1st approach

```python
grayLevel1 = cv.imread('g.jpg', 0)
[M, N] = grayLevel1.shape
for i in range(M):
  for j in range(N):
    if(grayLevel1[i,j] >= 100 and grayLevel1[i,j] <= 200):
      grayLevel1[i,j] = 255 #l-1 white
    else:
      grayLevel1[i,j] = 0

plt.figure(figsize=(20, 20))
plt.title(['Gray Level Slicing: 1st approach'], fontsize=30)
plt.imshow(grayLevel1, cmap='gray')
plt.axis('off')
plt.show(block=True)
```
