# Importing Libraries

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt 
from google.colab.patches import cv2_imshow
```

# Temporal Processing

#1. Adding two images

```python
rectangle = np.zeros((300,300),dtype="uint8")
cv.rectangle(rectangle,(50,50),(150,150),255,-1)
circle = 255 * np.zeros((300,300),dtype="uint8")
cv.circle(circle, center=(150,150), radius=50, color=(255,255,255), thickness=-1)

f = plt.figure(figsize=(25, 8))
f.add_subplot(1, 3, 1)
plt.title(['Rectangle'], fontsize=30)
plt.imshow(rectangle, cmap='gray')
plt.axis('off')
f.add_subplot(1, 3, 2)
plt.title(['Circle'], fontsize=30)
plt.imshow(circle, cmap='gray')
plt.axis('off')
f.add_subplot(1, 3, 3)
plt.title(['Rectangle + Circle'], fontsize=30)
plt.imshow(rectangle+circle, cmap='gray')
plt.axis('off')
plt.show(block=True)
```

```python
f = plt.figure(figsize=(25, 8))
f.add_subplot(1, 3, 1)
plt.title(['Rectangle'], fontsize=30)
plt.imshow(rectangle, cmap='gray')
plt.axis('off')
f.add_subplot(1, 3, 2)
plt.title(['Circle'], fontsize=30)
plt.imshow(circle, cmap='gray')
plt.axis('off')
f.add_subplot(1, 3, 3)
plt.title(['Rectangle - Circle'], fontsize=30)
plt.imshow(rectangle - circle, cmap='gray')
plt.axis('off')
plt.show(block=True)
```

```python
f = plt.figure(figsize=(25, 8))
f.add_subplot(1, 3, 1)
plt.title(['Rectangle'], fontsize=30)
plt.imshow(rectangle, cmap='gray')
plt.axis('off')
f.add_subplot(1, 3, 2)
plt.title(['Circle'], fontsize=30)
plt.imshow(circle, cmap='gray')
plt.axis('off')
f.add_subplot(1, 3, 3)
plt.title(['Rectangle * Circle'], fontsize=30)
plt.imshow(rectangle * circle, cmap='gray')
plt.axis('off')
plt.show(block=True)
```

# 4. Signal Averaging

```python
line = np.zeros((300, 300), dtype="uint8")
cv.line(line, (0, 0), (300, 300), color=(255,255,255))

f = plt.figure(figsize=(25, 8))
f.add_subplot(1, 4, 1)
plt.title(['Rectangle'], fontsize=30)
plt.imshow(rectangle, cmap='gray')
plt.axis('off')
f.add_subplot(1, 4, 2)
plt.title(['Circle'], fontsize=30)
plt.imshow(circle, cmap='gray')
plt.axis('off')
f.add_subplot(1, 4, 3)
plt.title(['Line'], fontsize=30)
plt.imshow(line, cmap='gray')
plt.axis('off')
f.add_subplot(1, 4, 4)
plt.title(['Signal Averaging'], fontsize=30)
plt.imshow((circle + rectangle + line)/3, cmap='gray')
plt.axis('off')
plt.show(block=True)
```

# 2. Rotating

```python
img = cv.imread('cat.jpg')
ht, wt, c = img.shape
Center_X, Center_Y = (wt//2, ht//2)
angle = 45
rot_matrix = cv.getRotationMatrix2D((Center_X, Center_Y), angle, 1.0)
rotate = cv.warpAffine(img, rot_matrix, (wt, ht))
f = plt.figure(figsize=(25, 8))
f.add_subplot(1, 2, 1)
plt.title(['Image 1'], fontsize=30)
plt.imshow(img, cmap='gray')
plt.axis('off')
f.add_subplot(1, 2, 2)
plt.title(['Roated Image'], fontsize=30)
plt.imshow(rotate, cmap='gray')
plt.axis('off')
plt.show(block=True)
```

# 3. Flipping

```python
img = cv.imread('cat.jpg')
Flipped = cv.flip(img,0)
f = plt.figure(figsize=(25, 8))
f.add_subplot(1, 2, 1)
plt.title(['Image 1'], fontsize=30)
plt.imshow(img, cmap='gray')
plt.axis('off')
f.add_subplot(1, 2, 2)
plt.title(['Flipped Image'], fontsize=30)
plt.imshow(Flipped, cmap='gray')
plt.axis('off')
plt.show(block=True)
```

# Zooming/Shrinking

```python
from scipy.ndimage import zoom

def clipped_zoom(img, zoom_factor, **kwargs):
    h, w = img.shape[:2]

    # For multichannel images we don't want to apply the zoom factor to the RGB
    # dimension, so instead we create a tuple of zoom factors, one per array
    # dimension, with 1's for any trailing dimensions after the width and height.
    zoom_tuple = (zoom_factor,) * 2 + (1,) * (img.ndim - 2)

    # Zooming out
    if zoom_factor < 1:

        # Bounding box of the zoomed-out image within the output array
        zh = int(np.round(h * zoom_factor))
        zw = int(np.round(w * zoom_factor))
        top = (h - zh) // 2
        left = (w - zw) // 2

        # Zero-padding
        out = np.zeros_like(img)
        out[top:top+zh, left:left+zw] = zoom(img, zoom_tuple, **kwargs)

    # Zooming in
    elif zoom_factor > 1:

        # Bounding box of the zoomed-in region within the input array
        zh = int(np.round(h / zoom_factor))
        zw = int(np.round(w / zoom_factor))
        top = (h - zh) // 2
        left = (w - zw) // 2

        out = zoom(img[top:top+zh, left:left+zw], zoom_tuple, **kwargs)

        # `out` might still be slightly larger than `img` due to rounding, so
        # trim off any extra pixels at the edges
        trim_top = ((out.shape[0] - h) // 2)
        trim_left = ((out.shape[1] - w) // 2)
        out = out[trim_top:trim_top+h, trim_left:trim_left+w]

    # If zoom_factor == 1, just return the input array
    else:
        out = img
    return out
```

```python
zm1 = clipped_zoom(img, 0.5)
zm2 = clipped_zoom(img, 1.5)
f = plt.figure(figsize=(25, 8))
f.add_subplot(1, 3, 1)
plt.title(['Original Image'], fontsize=30)
plt.imshow(img)
plt.axis('off')
f.add_subplot(1, 3, 2)
plt.title(['Shrinked Image'], fontsize=30)
plt.imshow(zm1)
plt.axis('off')
f.add_subplot(1, 3, 3)
plt.title(['Zoomed Image'], fontsize=30)
plt.imshow(zm2)
plt.axis('off')
plt.show(block=True)
```
