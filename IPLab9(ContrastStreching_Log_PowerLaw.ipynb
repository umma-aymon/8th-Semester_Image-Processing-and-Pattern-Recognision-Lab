# 1. Contrast Streching Method 1

```python
img = cv.imread('cat.jpg',0)
[M,N] = img.shape
T1=28
T2=75
for i in range (M):
 for j in range (N):
  if (img[i,j]>=T1) & (img[i,j]<=T2):
   img[i,j]=(227 * img[i,j] - 5040) / 47
  else:
   pass
res = np.hstack([im, img])
f=plt.figure(figsize=(25,8))
plt.title('Original Image vs Contrast Streching Method 1', fontsize=30, color='black', pad=20)
plt.imshow(res, cmap='gray')
plt.axis('off')
plt.show(block=True)
```

# 2. Contrast Streching Method 2

```python
T1=90
T2=138
img1 = cv.imread('cat.jpg',0)
for i in range (M):
 for j in range (N):
  if (img1[i,j]<T1):
   img1[i,j]= 0
  elif (img1[i,j]>=T1) & (img1[i,j]<=T2):
   img1[i,j]=(255 * img1[i,j] -22950)/48
  elif (img1[i,j]>T2):
   img1[i,j]=255
res = np.hstack([im, img1])
f=plt.figure(figsize=(25,8))
plt.title('Original Image vs Contrast Streching Method 2', fontsize=30, color='black', pad=20)
plt.imshow(res, cmap='gray')
plt.axis('off')
plt.show(block=True)
```

# 3. Log Transformation

```python
img = cv.imread('cat.jpg',0)
c = 255 / np.log(1 + np.max(img))
log_image = c * (np.log(img + 1))
log_image = np.array(log_image, dtype = np.uint8)
res = np.hstack([img, log_image])
f=plt.figure(figsize=(25,8))
plt.title('Original Image vs Log Transformation', fontsize=30, color='black', pad=20)
plt.imshow(res, cmap='gray')
plt.axis('off')
plt.show(block=True)
```

# 4. Power-Law transformation

```python
img = cv.imread('cat.jpg',0)
[M,N] = img.shape
x=[]
y=[]
c = 255 / np.log(1 + np.max(img))
for i in range (M):
 for j in range (N):
  img[i,j] = c * pow((img[i,j]),(1/2.5)) #gamma = 0.4
  x.append(img[i,j])
  y.append(img[i,j])
res = np.hstack([im, img])
f=plt.figure(figsize=(25,8))
plt.title('Original Image vs Power-Law Transformation', fontsize=30, color='black', pad=20)
plt.imshow(res, cmap='gray')
plt.axis('off')
plt.show(block=True)
```
